### 1.You’re given:

- A sorted array (this is VERY important)
- A target sum

You need to find two numbers whose sum = target

array = [2, 7, 11, 15], target = 9, output = [1, 2]

```python
def twoSumTarget(n, target):
    n.sort()
    low = 0
    high = len(n) - 1
    
    while low < high:
        currentSum = n[low] + n[high]
        if currentSum == target:
            return [n[low], n[high]]
        elif currentSum > target:
            high -= 1
        elif currentSum < target:
            low += 1
    
    return None

print(twoSumTarget([2,7,11,15], 9))
```

### 2. Remove Duplicates from Sorted Array

Your task is to remove the duplicates in-place such that each unique element appears only once.

- The relative order of the elements should be kept the same.
- Return the number of unique elements k.
- Modify the array such that the first k elements contain the unique elements.
- The remaining elements beyond k are not important.

nums = [1, 1, 2], Output: 2, Modified nums: [1, 2, _]

```python
def removeDuplicate(n):
    if len(n) <= 1: return n
    unique = 0
    
    
    for next in range(1, len(n)):
        if n[next] != n[unique]:
            unique += 1
            n[unique] = n[next]
            
    return n[:unique + 1]

print(removeDuplicate([1,1,2]))
```

### 3. Given an integer array nums and an integer `val`, remove all occurrences of val in-place. The relative order of the elements may be changed. After removing the elements, return the number of elements that are not equal to val.

Consider the number of elements not equal to val as k. To be accepted, you need to do the following:

* Modify the array nums such that the first k elements contain the elements not equal to val.
* The remaining elements of nums are not important.
* Return k.

nums = [3,2,2,3], val = 3, output = 2

nums = [0,1,2,2,3,0,4,2], val = 2, output = 5

```python
def removeVal(n, val):
    first = 0
    last = len(n) - 1

    while first <= last:

        if n[first] == val:
            while first <= last and n[last] == val:
                last -= 1
            
            if first <= last:
                n[first], n[last] = n[last], n[first]
                last -= 1
        else:
            first += 1

    return n[:last + 1]


print(removeVal([0,1,2,2,3,0,4,2], 2))
```

### 4. Given an integer array nums, move all 0s to the end of the array while maintaining the relative order of the non-zero elements.

* You must do this in-place (without making a copy of the array).
* Minimise the number of operations.

Input:  nums = [0,1,0,3,12], Output: [1,3,12,0,0]

```python
def moveZerosToEnd(n):
    slow = 0
    
    for next in range(len(n)):
        if n[next] != 0:
            n[next], n[slow] = n[slow], n[next]
            slow += 1
    return n

print(moveZerosToEnd([0,1,0,3,12]))
```

```python
def moveZerosToFront(n):
    slow = len(n) - 1
    
    for next in range(len(n) - 1, -1, -1):
        if n[next] != 0:
            n[next], n[slow] = n[slow], n[next]
            slow -= 1
    return n

print(moveZerosToFront([0,1,0,3,12]))
print(moveZerosToFront([4,0,5,0,3]))
```

### 5. Given a string s, determine if it is a palindrome, considering only alphanumeric characters and ignoring cases.

A string is a palindrome if it reads the same forward and backward after removing all non-alphanumeric characters and converting all letters to lowercase.

Return true if the string is a palindrome, otherwise return false.

Constraints:

* 1 <= s.length <= 2 * 10^5
* s consists only of printable ASCII characters.

Examples:

* Input: "A man, a plan, a canal: Panama" → Output: true
* Input: "race a car" → Output: false
* Input: " " → Output: true

```python
def isValidPalindrome(s):
    if len(s) == 1:
        return True
    
    left = 0
    right = len(s) - 1
    
    while left < right:
        while left < right and not s[left].isalnum():
            left += 1
        while left < right and not s[right].isalnum():
            right -= 1
        if s[left].lower() != s[right].lower():
            return False
        left += 1
        right -= 1
        
    return True

print(isValidPalindrome("A man, a plan, a canal: Panama")) # true
print(isValidPalindrome("race a car")) # false
print(isValidPalindrome(" ")) # true
```

### 6. Given an array of characters s, reverse the array in-place such that the order of the characters is reversed.

* You must modify the input array directly.

* Do not allocate extra space for another array (i.e., solve with O(1) extra memory).

Input: s = ["h","e","l","l","o"]

Output: ["o","l","l","e","h"]

```python
def reverseCharacters(n):
    if len(n) <= 1:
        return n
    
    head = 0
    tail = len(n) - 1
    
    while head < tail:
        n[head], n[tail] = n[tail], n[head]
        head += 1
        tail -= 1

array = ["h","e","l","l","o"]
reverseCharacters(["h","e","l","l","o"])
print(array)
```

### 7. Given an integer array nums sorted in non-decreasing order, return an array of the squares of each number, also sorted in non-decreasing order.

Constraints:

- 1 <= nums.length <= 10^5

- -10^4 <= nums[i] <= 10^4

- nums is sorted in non-decreasing order

```python
def sortSquaredArray(nums):
    n = len(nums)
    result = [0] * n
    
    left = 0
    right = n - 1
    position = n - 1
    
    while left <= right:
        if abs(nums[left]) > abs(nums[right]):
            result[position] = nums[left] * nums[left]
            left += 1
        else:
            result[position] = nums[right] * nums[right]
            right -= 1
        position -= 1
        
    return result

print(sortSquaredArray([-7, -3, 2, 3, 11]))
print(sortSquaredArray([-4, -1, 0, 3, 10]))
```

### 8. Given an integer array nums, return all the unique triplets [nums[i], nums[j], nums[k]] such that:

- i != j, i != k, and j != k

- nums[i] + nums[j] + nums[k] == 0

Requirements

- The solution set must not contain duplicate triplets.

- You can return the triplets in any order.

Input:

nums = [-1, 0, 1, 2, -1, -4]

Output:
[[-1, -1, 2], [-1, 0, 1]]

```python
def uniqueTripletsThreeSum(nums):
    nums.sort()
    triplets = []

    for i in range(len(nums) - 2):
        # Skip duplicate values for i
        if i > 0 and nums[i] == nums[i - 1]:
            continue

        left = i + 1
        right = len(nums) - 1

        while left < right:
            print(i, left, right)
            total = nums[i] + nums[left] + nums[right]

            if total < 0:
                left += 1
            elif total > 0:
                right -= 1
            else:
                triplets.append([nums[i], nums[left], nums[right]])

                # Skip duplicates for left
                while left < right and nums[left] == nums[left + 1]:
                    left += 1

                # Skip duplicates for right
                while left < right and nums[right] == nums[right - 1]:
                    right -= 1

                left += 1
                right -= 1

    return triplets


print(uniqueTripletsThreeSum([-1, 0, 1, 2, -1, -4]))
```

### 9. Given an integer array nums of length n and an integer target, find three integers in nums such that the sum is closest to target.

Return the sum of the three integers.

You may assume that each input would have exactly one solution.

Constraints

- 3 ≤ nums.length ≤ 10^3

- -10^3 ≤ nums[i] ≤ 10^3

- -10^4 ≤ target ≤ 10^4

Input: nums = [-1, 2, 1, -4], target = 1

Output: 2

```python
def sumThreeNumbers(nums, target):
    nums.sort()
    closestSum = nums[0] + nums[1] + nums[2]

    for i in range(len(nums) - 2):
        left = i + 1
        right = len(nums) - 1

        while left < right:
            total = nums[i] + nums[left] + nums[right]

            if abs(total - target) < abs(closestSum - target):
                closestSum = total

            if total < target:
                left += 1
            elif total > target:
                right -= 1
            else:
                return total  # exact match

    return closestSum
    
print(sumThreeNumbers([-1, 2, 1, -4], 1))
print(sumThreeNumbers([1, 1, 2, 3], 4))
print(sumThreeNumbers([2, 2, 2, 2], 7))
print(sumThreeNumbers([-2, 0, 1, 1, 2], 3))
print(sumThreeNumbers([-1, -1, 2, 5], 1))
```

### 11. You are given an integer array height of length n. There are n vertical lines drawn such that the two endpoints of the i-th line are (i, 0) and (i, height[i]).

Find two lines that, together with the x-axis, form a container such that the container holds the maximum amount of water.

Return the maximum amount of water a container can store.

Note:

- You may not slant the container.
- The amount of water a container can store is determined by the shorter of the two lines and the distance between them.

Example 1:

Input: height = [1,8,6,2,5,4,8,3,7]

Output: 49

Example 2:

Input: height = [1,1]

Output: 1

Constraints:

- n == height.length
- 2 <= n <= 10^5
- 0 <= height[i] <= 10^4

```python
def maxWaterContainer(n):
    head = 0
    tail = len(n) - 1
    maxArea = 0
    
    while head < tail:
        area = min(n[head], n[tail]) * abs(tail - head)
        if maxArea < area:
            maxArea = area
        
        if n[head] > n[tail]:
            tail -= 1
        else:
            head += 1
        
    return maxArea

print(maxWaterContainer([1,8,6,2,5,4,8,3,7]))
```
